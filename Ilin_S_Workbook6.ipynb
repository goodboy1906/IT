### 1.1.1 –ü—Ä–∏–º–µ—Ä

–ó–∞–¥–∞—á–∞:

        –ü—É—Å—Ç—å –¥–∞–Ω–∞ –Ω–∞—á–∞–ª—å–Ω–∞—è –ø–æ–ø—É–ª—è—Ü–∏—è –∏–∑ —á–µ—Ç—ã—Ä–µ—Ö —Ö—Ä–æ–º–æ—Å–æ–º —Å –¥–≤—É–º—è–≥–µ–Ω–∞–º–∏ x –∏ y. –ü–æ–∫–∞–∑–∞—Ç–µ–ª—å –∫–∞—á–µ—Å—Ç–≤–∞ —Ö—Ä–æ–º–æ—Å–æ–º—ã –æ—Ü–µ–Ω–∏–≤–∞–µ—Ç—Å—è —Ñ—É–Ω–∫—Ü–∏–µ–π Z.–ü—Ä–∏ —Ä–∞–≤–Ω–æ–º –∫–∞—á–µ—Å—Ç–≤–µ —Ö—Ä–æ–º–æ—Å–æ–º –ø—Ä–µ–¥–ø–æ—á—Ç–µ–Ω–∏–µ –æ—Ç–¥–∞–µ—Ç—Å—è —Ö—Ä–æ–º–æ—Å–æ–º–µ —Å–±–æ–ª—å—à–∏–º –Ω–æ–º–µ—Ä–æ–º. –ù–∞ –∫–∞–∂–¥–æ–º —ç—Ç–∞–ø–µ —Ö—Ä–æ–º–æ—Å–æ–º–∞ a —Å –≤—ã—Å—à–∏–º –∫–∞—á–µ—Å—Ç–≤–æ–º–ø–æ—Ä–æ–∂–¥–∞–µ—Ç —á–µ—Ç—ã—Ä–µ –Ω–æ–≤—ã—Ö —Ö—Ä–æ–º–æ—Å–æ–º—ã ùëè1, ùëê1, ùëè2, ùëê2, –æ–±–º–µ–Ω–∏–≤–∞—è—Å—å –≥–µ–Ω–∞–º–∏ —Å–¥–≤—É–º—è —Ö—Ä–æ–º–æ—Å–æ–º–∞–º–∏ b –∏ c –±–æ–ª–µ–µ –Ω–∏–∑–∫–æ–≥–æ –∫–∞—á–µ—Å—Ç–≤–∞ –ø–æ —É–∫–∞–∑–∞–Ω–Ω–æ–π —Å—Ö–µ–º–µ:


        –ü–æ—Å–ª–µ–¥–Ω—è—è —Ö—Ä–æ–º–æc–æ–º–∞ (—Å –Ω–∏–∑—à–∏–º –∫–∞—á–µ—Å—Ç–≤–æ–º) –≤—ã–±—ã–≤–∞–µ—Ç –∏–∑–ø–æ–ø—É–ª—è—Ü–∏–∏. –ù–∞–π—Ç–∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å –∫–∞—á–µ—Å—Ç–≤–∞ —Ö—Ä–æ–º–æ—Å–æ–º—ã –≤–ø–æ–ø—É–ª—è—Ü–∏–∏ –∏ –æ–±—â–µ–µ –∫–∞—á–µ—Å—Ç–≤–æ –ø–æ–ø—É–ª—è—Ü–∏–∏ –ø–æ—Å–ª–µ —á–µ—Ç—ã—Ä–µ—Ö —ç—Ç–∞–ø–æ–≤ —ç–≤–æ–ª—é—Ü–∏–∏.
        
        –ü–æ—Ç—Ä–µ–±—É–µ—Ç—Å—è –Ω–µ—Å–∫–æ–ª—å–∫–æ —Ñ—É–Ω–∫—Ü–∏–π –¥–ª—è —Ä–µ–∞–ª–∏–∑–∞—Ü–∏–∏ –∞–ª–≥–æ—Ä–∏—Ç–º–∞. –ù–∞–ø–∏—à–µ–º –∏—Ö.
        
        –†–µ—à–µ–Ω–∏–µ: –ù–∞—á–Ω–µ–º —Å —Ñ—É–Ω–∫—Ü–∏–∏ –æ—Ü–µ–Ω–∫–∏ –∫–∞—á–µ—Å—Ç–≤–∞ —Ö—Ä–æ–º–æ—Å–æ–º—ã qZ(x,y):

```python
# —Ñ—É–Ω–∫—Ü–∏—è –∫–∞—á–µ—Å—Ç–≤–∞ —Ö—Ä–æ–º–æ—Å–æ–º—ã 
def qZ(x, y):
    return (x - 3 * y + 1) / (3* x ** 2 + 3 * y ** 2 + 1)
```

        –î–∞–ª–µ–µ, –æ—Ü–µ–Ω–∏–º —Å—É–º–º–∞—Ä–Ω–æ–µ –∫–∞—á–µ—Å—Ç–≤–æ —Ö—Ä–æ–º–æ—Å–æ–º:

```python
# —Å—É–º–º–∞ –∫–∞—á–µ—Å—Ç–≤–∞ —Ö—Ä–æ–º–æ—Å–æ–º 
def qSumZ(Z):
    return sum(Z)
```

        –ò –∑–∞–ø—Ä–æ–≥—Ä–∞–º–º–∏—Ä—É–µ–º –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–Ω—É—é –≤—ã—à–µ —Å—Ö–µ–º—É –æ–±–º–µ–Ω–∞ —Ö—Ä–æ–º–æ—Å–æ–º–∞–º–∏:

```python
def exchangeScheme(oldX, oldY, sortedId):
    X = [0 for i in range(4)] 
    Y = [0 for i in range(4)]

    X[2] = oldX[sortedId[2]] 
    X[3] = oldX[sortedId[2]]

    X[0] = oldX[sortedId[0]]

    X[1] = oldX[sortedId[1]]

    Y[0] = oldY[sortedId[2]] 
    Y[1] = oldY[sortedId[2]]

    Y[2] = oldY[sortedId[0]]

    Y[3] = oldY[sortedId[1]]

    return X, Y
```

        –û—Ç—Å–æ—Ä—Ç–∏—Ä—É–µ–º –º–∞—Å—Å–∏–≤ –∫–∞—á–µ—Å—Ç–≤–∞ –Ω–∞—à–∏—Ö –ø–æ—Ç–æ–º–∫–æ–≤ –∏ –≤—ã–¥–µ–ª–∏–º –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ –∏–Ω–¥–µ–∫—Å—ã:

```python
def sorting(Z):
    sortedId = sorted(range(len(Z)), key = lambda k: Z[k])
    return sortedId
```

        –ù–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è —à–∞–≥–∞ —ç–≤–æ–ª—é—Ü–∏–∏:

```python
# —à–∞–≥ —ç–≤–æ–ª—é—Ü–∏–∏ 
def evoStep(X, Y, Z):
    _, minId = min((value, id) for (id, value) in enumerate(Z))
    X = X[:] 
    Y = Y[:] 
    Z = Z[:]
    
    X.pop(minId)
    Y.pop(minId)
    Z.pop(minId)
    
    return X, Y, Z
```

        –ü—Ä–æ–∏–∑–≤–µ–¥–µ–º —ç–≤–æ–ª—é—Ü–∏–æ–Ω–Ω—ã–µ –∏–∑–º–µ–Ω–µ–Ω–∏—è, –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å –∑–∞–¥–∞—á–µ–π - 4 —à–∞–≥–∞:

```python
# —à–∞–≥–∏ —ç–≤–æ–ª—é—Ü–∏–∏ (–∫–æ–Ω–µ—á–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è), –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é 4 —à–∞–≥–∞ 
def evoSteps(X, Y, stepsNum = 4):
    results = []
    
    for i in range(4):
        arrZ = [qZ(x, Y[i]) for i, x in enumerate(X)]
        
        X, Y, Z = evoStep(X, Y, arrZ)
        
        X, Y = exchangeScheme(X, Y, sorting(Z))
        
        results.append([X, Y, qSumZ(arrZ), arrZ])
        
    return X, Y, results
```

        –¢–µ–ø–µ—Ä—å, –∫–æ–≥–¥–∞ –º—ã –ø–æ–¥–≥–æ—Ç–æ–≤–∏–ª–∏—Å—å –∫ —Ä–µ—à–µ–Ω–∏—é –∑–∞–¥–∞—á–∏, –Ω–∞–ø–∏—Å–∞–≤ –≤—Å–µ–Ω–µ–æ–±—Ö–æ–¥–∏–º—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è —Ä–µ–∞–ª–∏–∑–∞—Ü–∏–∏ –≥–µ–Ω–µ—Ç–∏—á–µ—Å–∫–æ–≥–æ –∞–ª–≥–æ—Ä–∏—Ç–º–∞ (–æ—Ü–µ–Ω–∫–∏ –∫–∞—á–µ—Å—Ç–≤–∞ —Ö—Ä–æ–º–æ—Å–æ–º, —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ –ø–æ—Ç–æ–º–∫–æ–≤ –∏ —ç–≤–æ–ª—é—Ü–∏–æ–Ω–Ω—ã—Ö —à–∞–≥–æ–≤), —Ä–µ—à–∏–º–∑–∞–¥–∞—á—É –≤ —á–∏—Å–ª–∞—Ö. –ü—É—Å—Ç—å –¥–∞–Ω—ã —Å–ª–µ–¥—É—é—â–∏–µ –º–∞—Å—Å–∏–≤—ã —Ö—Ä–æ–º–æ—Å–æ–º X –∏ Y:


        –ó–∞–ø–∏—à–µ–º –∏—Ö –≤ —Ç—Ä–µ–±—É–µ–º–æ–º –≤–∏–¥–µ –∏ –≤–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è –Ω–∞–ø–∏—Å–∞–Ω–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–µ–π evoSteps.

```python
# –æ–±—ä—è–≤–ª–µ–Ω–∏–µ –º–∞—Å—Å–∏–≤–æ–≤ —Ö—Ä–æ–º–æ—Å–æ–º
X = [-2, -1, 0, 1] 
Y = [-2, -1, 0, 1]
```

```python
# –†–µ–∞–ª–∏–∑–∞—Ü–∏—è –∞–ª–≥–æ—Ä–∏—Ç–º–∞ 
results = evoSteps(X, Y)
```

        –¢–µ–ø–µ—Ä—å, –≤—ã–≤–µ–¥–µ–º –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –¥–ª—è –ø–æ–∫–∞–∑–∞—Ç–µ–ª—è –∫–∞—á–µ—Å—Ç–≤–∞ —Ö—Ä–æ–º–æ—Å–æ–º—ã–≤ –ø–æ–ø—É–ª—è—Ü–∏–∏ –∏ –æ–±—â–µ–µ –∫–∞—á–µ—Å—Ç–≤–æ –ø–æ–ø—É–ª—è—Ü–∏–∏ –ø–æ—Å–ª–µ —á–µ—Ç—ã—Ä–µ—Ö —ç—Ç–∞–ø–æ–≤ —ç–≤–æ–ª—é—Ü–∏–∏. –î–ª—è —ç—Ç–æ–≥–æ, –≤–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è —Ü–∏–∫–ª–æ–º –ø–æ –∑–Ω–∞—á–µ–Ω–∏—è–º –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π results.

```python
for i in range(len(results[2])):
    print(f'max_{i + 1}_step: {results[2][i][2]}')
    
qualityArrZ = [] 
for i in range(len(results[2])):
    qualityArrZ += results[2][i][3]
    
print(f'max Z: {max(qualityArrZ)}')
```

```python
def qZ(x,y):
    return (x-3*y)/(2*x**2+2*y**2+1)

def qSumZ(Z):
    return sum(Z)

def exchangeScheme(oldX, oldY, sortedId):
    X = [0 for i in range(4)]
    Y = [0 for i in range(4)]
    
    X[2] = oldX[sortedId[2]]
    X[3] = oldX[sortedId[2]]
    
    X[0] = oldX[sortedId[0]]
    
    X[1] = oldX[sortedId[1]]
    
    Y[0] = oldX[sortedId[2]]
    Y[1] = oldX[sortedId[2]]
    
    Y[2] = oldX[sortedId[0]]
    
    Y[3] = oldX[sortedId[1]]
    
    return X, Y

def sorting(Z):
    sortedId = sorted(range(len(Z)), key = lambda k: Z[k])
    
    return sortedId

def evoStep(X, Y, Z):
    _, minId = min((value, id) for (id, value) in enumerate(Z))
    X = X[:]
    Y = Y[:]
    Z = Z[:]
    
    X.pop(minId)
    Y.pop(minId)
    Z.pop(minId)
    
    return X, Y, Z

def evoSteps(X, Y, stepsNum = 4):
    results = []
    
    for i in range(4):
        arrZ = [qZ(x, Y[i]) for i, x in enumerate(X)]
        
        X, Y, Z = evoStep(X, Y, arrZ)
        
        X, Y = exchangeScheme(X, Y, sorting(Z))
        
        results.append([X, Y, qSumZ(arrZ), arrZ])
        
    return X, Y, results

X = [-1, 0, 2, 3]
Y = [0, 1, -2, 2]

results = evoSteps(X, Y)

for i in range(len(results[2])):
    print(f'max_{i + 1}_step: {results[2][i][2]}')
    
qualityArrZ = []
for i in range(len(results[2])):
    qualityArrZ += results[2][i][3]
    
print(f'max Z: {max(qualityArrZ)}')
```

```python
distances = [(1, 2, 20), 
             (1, 3, 40), 
             (1, 4, 42), 
             (1, 5, 33), 
             (1, 6, 21), 
             (2, 3, 26), 
             (2, 4, 38), 
             (2, 5, 42), 
             (2, 6, 17), 
             (3, 4, 22), 
             (3, 5, 43), 
             (3, 6, 21), 
             (4, 5, 27), 
             (4, 6, 22), 
             (5, 6, 26)] # –¥–ª–∏–Ω—ã —Ä—ë–±–µ—Ä

V = [1, 4, 5, 2, 6, 3, 1] # –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –ø—Ä–æ—Ö–æ–∂–¥–µ–Ω–∏—è –º–∞—Ä—à—Ä—É—Ç–∞ 
Z = [(3, 4), 
     (4, 6), 
     (5, 2), 
     (6, 2)] # –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∑–∞–º–µ–Ω –≤–µ—Ä—à–∏–Ω
P = [49, 54, 43, 54] # —Å–ª—É—á–∞–π–Ω—ã–µ —á–∏—Å–ª–∞, –≤—ã–ø–∞–≤—à–∏–µ –≤ –ø—Ä–æ—Ü–µ—Å—Å–µ —Å—á—ë—Ç–∞

T = 100 # –Ω–∞—á–∞–ª—å–Ω–∞—è —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞
```

        –ó–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –∏ –∏–∑–º–µ–Ω–µ–Ω–∏—è —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä—ã:

```python
# —Ñ—É–Ω–∫—Ü–∏—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ 
def probability(delta, T):
    return 100 * e ** (-delta / T)

# —Ñ—É–Ω–∫—Ü–∏—è –∏–∑–º–µ–Ω–µ–Ω–∏—è —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä—ã 
def reductTemp(prevT):
    nextT = 0.5 * prevT

    return nextT
```

        –ü–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ –ø–æ –∑–∞–¥–∞–Ω–Ω—ã–º –≤–µ—Ä—à–∏–Ω–∞–º, –∫–∞–∫ –≤ –Ω–∞—á–∞–ª—å–Ω–æ–º —É—Å–ª–æ–≤–∏–∏:

```python
graph = nx.Graph() # —Å–æ–∑–¥–∞–Ω–∏–µ –ø—É—Å—Ç–æ–≥–æ –≥—Ä–∞—Ñ–∞ 
graph.add_weighted_edges_from(distances) # –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –±–µ—Å–æ–≤ —Ä—ë–±–µ—Ä

# –æ—Ç—Ä–∏—Å–æ–≤–∫–∞ –≥—Ä–∞—Ñ–∞ —Å –∑–∞–¥–∞–Ω–Ω—ã–º–∏ –≤–µ—Ä—à–∏–Ω–∞–º–∏ 
nx.draw_kamada_kawai(graph, node_color = '#fb7258', node_size = 2000, with_labels = True)
```

        –¢–µ–ø–µ—Ä—å, –Ω–∞–ø–∏—à–µ–º –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã–µ, —Å–æ–≥–ª–∞—Å–Ω–æ –∞–ª–≥–æ—Ä–∏—Ç–º—É –≤—ã—à–µ, —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è —Ä–∞–±–æ—Ç—ã –º–µ—Ç–æ–¥–∞ –∏–º–∏—Ç–∞—Ü–∏–∏ –æ—Ç–∂–∏–≥–∞. –í—ã—á–∏—Å–ª–∏–º –¥–ª–∏–Ω—É —Ä–µ–±—Ä–∞:

```python
# –±—ã—á–∏—Å–ª–µ–Ω–∏–µ –¥–ª–∏–Ω—ã —Ä–µ–±—Ä–∞ 
def edgeLength(i, j, distances, roundTrip = True):
    if roundTrip:
        return max([(item[2] if (item[0] -- i and item[1] -- j) or (item[1] -- i and item[0] -- j) else -1)
                    for item in distances])
    else: 
        return max([(item[2] if (item[0] == i and item[1] == j) else -1) for item in distances])
```

        –í—ã—á–∏—Å–ª–∏–º –¥–ª–∏–Ω—É –º–∞—Ä—à—Ä—É—Ç–∞:

```python
# –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ –¥–ª–∏–Ω—ã –º–∞—Ä—à—Ä—É—Ç–∞ 
def routeLength(V, distances):
    edges = []
    
    for i in range(len(V) - 1):
        edges.append(edgeLength(V[i], V[i + 1], distances))

    return sum(edges)
```

```python
# –æ–¥–Ω–∞ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∞ –≤ –ø—É—Ç–∏ 
def routeOneReplacement(arrV, Z, replacementByName = True):
    decrement = 1 if replacementByName else 0
    
    arrV[Z[0] - decrement], arrV[Z[1] - decrement] = arrV[Z[1] - decrement], arrV[Z[0] - decrement]
    
    return arrV
```

```python
# –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏ –≤ –ø—É—Ç–∏ 
def routeReplacement(V, Z):
    for z in Z:
        V = routeOneReplacement(V, z)
    return V
```

```python
# –±—ã–±–æ—Ä –Ω—É–∂–Ω–æ–≥–æ –ø—É—Ç–∏ –º–µ—Ç–æ–¥–æ–º –æ—Ç–∂–∏–≥–∞ 
def chooseRoute(distances, V, z, T, P):
    sumLength = routeLength(V, distances) # –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏–µ –¥–ª–∏–Ω—ã –ø—É—Ç–∏ 
    arrSum = [sumLength] # –º–∞—Å—Å–∏–≤ —Å—É–º–º –¥–ª–∏–Ω
# —Ü–∏–∫–ª—ã –º–µ—Ç–æ–¥–æ–º –æ–ø–∂–∏–≥–∞ 
    for i in range(len(Z)):
        newV = routeOneReplacement(V[:], Z[i]) # –Ω–æ–≤—ã–π –º–∞—Ä—à—Ä—É—Ç –ø–æ—Å–ª–µ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏ 
        newS = routeLength(newV, distances) # –¥–ª–∏–Ω–∞ –Ω–æ–≤–æ–≥–æ –º–∞—Ä—à—Ä—É—Ç–∞ 
        arrSum.append(newS)
        deltaS = newS - sumLength # —Ä–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É –¥–ª–∏–Ω–æ–π –Ω–æ–≤–æ–≥–æ –∏ —Å—Ç–∞—Ä–æ–≥–æ –º–∞—Ä—à—Ä—É—Ç–æ–≤

# –≤ —Å–ª—É—á–∞–µ, –µ—Å–ª–∏ —Ä–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É –¥–ª–∏–Ω–∞–º–∏ –±–æ–ª—å—à–µ –≤, —Ç–æ –≤—ã—á–∏—Å–ª—è–µ—Ç—Å—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å
        if deltaS > 0:
            p = probability(deltaS, T) # –ø–æ–¥—Å—á—ë—Ç –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏

# –µ—Å–ª–∏ –∑–∞–¥–∞–Ω–Ω–∞—è –±–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞–¥–∞–µ—Ç –± –∏–Ω—Ç–µ—Ä–±–∞–ª –æ—Ç 0 –¥–æ —Ä, —Ç–æ –Ω–æ–±—ã–π –º–∞—Ä—à—Ä—É—Ç –±—ã–±–∏—Ä–∞–µ—Ç—Å—è
            if p > P[i]:
                V = newV
                sumLength = newS
        else:
            V = newV
            sumLength = newS
            
        T = reductTemp(T) # –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä—ã

    return V, arrSum
```
